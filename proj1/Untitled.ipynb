```python
def text_to_hex(string):
    #returns a string that will be interpreted as the text entered by the whatever
    letters = []
    for l in string:
        letters.append(to_hex(ord(l)))
    return ''.join(letters)
    
    
def to_hex(num):
    return hex(num)[2:]

def to_int(num):
    return int(num, 16)

def hex_to_text(string):
    lets = []
    for x in string:
        lets.append(str(chr(to_int(x))))
    return ''.join(lets)
```
